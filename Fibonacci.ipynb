```python
import time
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
def Fibo(nterms):
    start_time = time.time()
    n1 = 0
    n2 = 1
    cont = 0
    if(nterms==1):
        print(n1)
    else: 
        while cont < nterms:
            #print(n1,end=', ')
            nth = n1 + n2
            n1 = n2
            n2 = nth
            cont += 1
            
    return (time.time() - start_time)
```

```python
nterms = int(input("Ingrese el n término "))
tiempo = []
for i in range(nterms):
    tiempo.append(Fibo(i)) 
```

```python
plt.ylabel("Tiempo (s)")
plt.xlabel('Tamaño (n)')
x = np.arange(nterms)
y = np.array(tiempo)
plt.plot(x, y,'k.')
```
